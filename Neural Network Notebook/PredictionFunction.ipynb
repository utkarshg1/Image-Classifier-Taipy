# Prediction of images function

```python
from keras.models import load_model
from PIL import Image
import numpy as np
```

```python
class_names = {
    0: 'airplane',
    1: 'automobile',
    2: 'bird',
    3: 'cat',
    4: 'deer',
    5: 'dog',
    6: 'frog',
    7: 'horse',
    8: 'ship',
    9: 'truck',
}
```

```python
model = load_model('Cifar10Model.keras')
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
def predict_image(model, file_path):
    img = Image.open(file_path)
    img = img.convert("RGB")
    print('Image Input : ')
    plt.imshow(img)
    plt.show()
    img = img.resize((32, 32))
    print('\n==============================================\n')
    print('Image after Resizing : ')
    plt.imshow(img)
    plt.show()
    data = np.asarray(img)
    data = data/255
    probs = model.predict(np.array([data])[:1])
    top_probs = probs.max()
    pred = class_names[np.argmax(probs)]
    return top_probs, pred
```

```python
predict_image(model, file_path='./demo_images/demo_img3.png')
```
